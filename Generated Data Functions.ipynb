The first csv file will contain employee id, age, efficiency, and salary. The second csv file will contain employee id, application totals for each month, annual average, and total application count. The third csv file will contain time used (in minutes) for the analyst to collect employee data for each month, annually, and total. The fourth csv file will contain time used (in minutes) for the analyst to perform data analysis and visualization for each month, annually, and total.The fifth csv file will contain time used (in minutes) for the analyst to collect employee data for each month, annually, and total with CRM assistance. The fourth csv file will contain time used (in minutes) for the analyst to perform data analysis and visualization for each month, annually, and total with CRM assistance.

```python
def employee_dta (numOfEmp):
    
    import pandas as pd
    import numpy as np
    
    emp_csv = 'emp'
    kpi_csv = 'kpi' 
    m_record_csv = 'm_record'
    m_analysis_csv = 'm_analysis' 
    a_record_csv = 'a_record'
    a_analysis_csv = 'a_analysis'
    
    
    '''
    ==================================================================
    CALCULATING EMP_ID & EMP_AGE 
    ==================================================================
    '''

    emp_id = np.arange(0, numOfEmp, 1)
    emp_age = np.random.randint(20, 56, size = numOfEmp)
    
    '''
    ==================================================================
    CALCULATING PRE-CRM EMP EFF & SAL
    Efficiency Limitation 
    Age 20-30(1-5%)
    Age 31+(1-2%)
    ==================================================================
    '''

    emp_eff = []

    for x in emp_age:
        if x >= 20 and x <= 30:
            y = np.random.randint(1,5)
            emp_eff.append(y * .01)
        else:
            y = np.random.randint(1,2)
            emp_eff.append(y * .01)
            
    emp_sal = []

    for x in emp_age:
        if x >= 20 and x <= 30:
            step = np.random.randint(1,5) * .01
            y = 35360 
            emp_sal.append(np.ceil((y + (y * step))).astype(int))     
        else:
            step = np.random.randint(1,5) * .01
            y = 45760
            emp_sal.append(np.ceil((y + (y * step))).astype(int)) 
            
    
    emp_data = {'employeeid': emp_id, 'age': emp_age, 'employeeeff': emp_eff, 'employeesal':emp_sal}
    emp_df = pd.DataFrame(data = emp_data)
    
    emp_df.to_csv(emp_csv + '.csv', index=False)
    
    '''
    ==================================================================
    
    CALCULATING PRE-CRM KPI

    Min/Max App can be done in a day = [480 minutes(8 hours a day) / (15 (low end) or 60 (high end)) 
    Min/Max App can be done in a year =  App Min/Max * 252 (Total amount of working days in a year)] 
    MinMax App / 12 = Lowest/Highest App amount per month
    ==================================================================

    '''
    
    kpi_data = ['employeeid','Jan', 'Feb', 'Mar', 'Apr', 
     'May', 'Jun', 'Jul', 'Aug',
     'Sep', 'Oct', 'Nov', 'Dec', 'Avg', 'Total', 'Category']

    kpi_months = ['Jan', 'Feb', 'Mar', 'Apr', 
     'May', 'Jun', 'Jul', 'Aug',
     'Sep', 'Oct', 'Nov', 'Dec']    
        
    kpi_df = pd.DataFrame(columns = kpi_data)
    kpi_months_tot = pd.DataFrame(columns = kpi_months)
    
    for x in range (len(emp_id)):
        applst = []

        for y in range (12):
            if y == 0:
                app = np.random.randint(168, 200)
                appeff = (app * emp_eff[x])
                tot1 = np.ceil(app + appeff).astype(int)
                applst.append(tot1)
            elif y == 1:
                app = np.random.randint(168, 357)
                appeff = (app * emp_eff[x])
                tot2 = np.ceil(app + appeff).astype(int)
                applst.append(tot2)
            else:
                app = np.random.randint(168, 672)
                appeff = (app * emp_eff[x])
                tot3 = np.ceil(app + appeff).astype(int)
                applst.append(tot3)

        average = np.ceil (sum(applst) / 12).astype(int)
        total = sum(applst)

        kpi_data = {'employeeid': emp_id[x],'Jan': applst[0] , 'Feb': applst[1], 'Mar': applst[2], 'Apr': applst[3], 
               'May': applst[4], 'Jun': applst[5], 'Jul': applst[6], 'Aug': applst[7],
               'Sep': applst[8], 'Oct': applst[9], 'Nov':applst[10], 'Dec': applst[11], 
               'Avg': average, 'Total': total, 'Category': 'PRE-CRM'}
        kpi_month_sums = {'Jan': applst[0] , 'Feb': applst[1], 'Mar': applst[2], 'Apr': applst[3], 
               'May': applst[4], 'Jun': applst[5], 'Jul': applst[6], 'Aug': applst[7],
               'Sep': applst[8], 'Oct': applst[9], 'Nov':applst[10], 'Dec': applst[11]}

        kpi_df = kpi_df.append(kpi_data, ignore_index = True)
        kpi_months_tot = kpi_months_tot.append(kpi_month_sums, ignore_index = True)

    kpi_df.to_csv(kpi_csv + '.csv', index=False)
    
    
    '''
    ==================================================================
    MANUAL DATA ANALYSIS

    ==================================================================

    '''
    
    monLst = ['Jan', 'Feb', 'Mar', 'Apr', 
     'May', 'Jun', 'Jul', 'Aug',
     'Sep', 'Oct', 'Nov', 'Dec', 'Annual']
    
    
    kpi_col = ['Jan', 'Feb', 'Mar', 'Apr', 
     'May', 'Jun', 'Jul', 'Aug',
     'Sep', 'Oct', 'Nov', 'Dec', 'Annual', 'Total']

    
    '''
    ==================================================================
    MANUAL RECORDING

    ==================================================================

    '''   
    m_recording_df = pd.DataFrame(columns = kpi_col)
    
    mapplst = []

    for x in monLst:
        if x != 'Annual':
            if  kpi_df[x].sum() <= 20000:
                emp_min = np.random.randint(60, 100)
                mapplst.append(emp_min)
            elif kpi_df[x].sum() > 20000 and kpi_df[x].sum() <= 30000:
                emp_min = np.random.randint(80, 120)
                mapplst.append(emp_min)
            else:
                emp_min = np.random.randint(100, 180)
                mapplst.append(emp_min)
        else:

            if  kpi_months_tot.sum().sum() <= 20000:
                emp_min = np.random.randint(60, 100)
                mapplst.append(emp_min)
            elif kpi_months_tot.sum().sum() > 20000 and kpi_months_tot.sum().sum() <= 30000:
                emp_min = np.random.randint(80, 120)
                mapplst.append(emp_min)
            else:
                emp_min = np.random.randint(100, 180)
                mapplst.append(emp_min)

        mtotal = sum(mapplst)

    mkpi_data = {'Jan': mapplst[0] , 'Feb': mapplst[1], 'Mar': mapplst[2], 'Apr': mapplst[3], 
               'May': mapplst[4], 'Jun': mapplst[5], 'Jul': mapplst[6], 'Aug': mapplst[7],
               'Sep': mapplst[8], 'Oct': mapplst[9], 'Nov': mapplst[10], 'Dec': mapplst[11], 
               'Annual': mapplst[12], 'Total': mtotal}

    m_recording_df = m_recording_df.append(mkpi_data, ignore_index = True)
    m_recording_df.to_csv(m_record_csv + '.csv', index=False)

    '''
    ==================================================================
    MANUAL ANALYSIS

    ==================================================================

    '''
    m_analsis_df = pd.DataFrame(columns = kpi_col)
    
    aapplst = []

    for x in monLst:
        if x != 'Annual':
            if  kpi_df[x].sum() <= 20000:
                emp_min = np.random.randint(120, 150)
                aapplst.append(emp_min)
            elif kpi_df[x].sum() > 20000 and kpi_df[x].sum() <= 30000:
                emp_min = np.random.randint(140, 200)
                aapplst.append(emp_min)
            else:
                emp_min = np.random.randint(180, 240)
                aapplst.append(emp_min)
        else:

            if  kpi_months_tot.sum().sum() <= 35000:
                emp_min = np.random.randint(240, 300)
                aapplst.append(emp_min)
            elif kpi_months_tot.sum().sum() > 35000 and kpi_months_tot.sum().sum() <= 400000:
                emp_min = np.random.randint(280, 310)
                aapplst.append(emp_min)
            else:
                emp_min = np.random.randint(300, 360)
                aapplst.append(emp_min)

        mtotal = sum(aapplst)

    mAna_data = {'Jan': aapplst[0] , 'Feb': aapplst[1], 'Mar': aapplst[2], 'Apr': aapplst[3], 
               'May': aapplst[4], 'Jun': aapplst[5], 'Jul': aapplst[6], 'Aug': aapplst[7],
               'Sep': aapplst[8], 'Oct': aapplst[9], 'Nov': aapplst[10], 'Dec': aapplst[11], 
               'Annual': aapplst[12], 'Total': mtotal}

    m_analsis_df =  m_analsis_df.append(mAna_data, ignore_index = True)
    m_analsis_df.to_csv(m_analysis_csv + '.csv', index=False)

    '''
    ==================================================================
    AUTOMATIC RECORDING

    ==================================================================

    '''
    
    a_recording_df = pd.DataFrame(columns = kpi_col)
    
    crmapplst = []

    for x in monLst:
        if x != 'Annual':
            if  kpi_df[x].sum() <= 20000:
                emp_min = np.random.randint(20, 30)
                crmapplst.append(emp_min)
            elif kpi_df[x].sum() > 20000 and kpi_df[x].sum() <= 30000:
                emp_min = np.random.randint(25, 35)
                crmapplst.append(emp_min)
            else:
                emp_min = np.random.randint(30, 45)
                crmapplst.append(emp_min)
        else:

            if  kpi_months_tot.sum().sum() <= 300000:
                emp_min = np.random.randint(15, 20)
                crmapplst.append(emp_min)
            elif kpi_months_tot.sum().sum() > 300000 and kpi_months_tot.sum().sum() <= 350000:
                emp_min = np.random.randint(20, 25)
                crmapplst.append(emp_min)
            else:
                emp_min = np.random.randint(25, 30)
                crmapplst.append(emp_min)

        mtotal = sum(crmapplst)

    crmkpi_data = {'Jan': crmapplst[0] , 'Feb': crmapplst[1], 'Mar': crmapplst[2], 'Apr': crmapplst[3], 
               'May': crmapplst[4], 'Jun': crmapplst[5], 'Jul': crmapplst[6], 'Aug': crmapplst[7],
               'Sep': crmapplst[8], 'Oct': crmapplst[9], 'Nov': crmapplst[10], 'Dec': crmapplst[11], 
               'Annual': crmapplst[12], 'Total': mtotal}

    a_recording_df = a_recording_df.append(crmkpi_data, ignore_index = True)
    a_recording_df.to_csv(a_record_csv + '.csv', index=False)

    '''
    ==================================================================
    AUTOMATIC RECORDING

    ==================================================================

    '''
    
    a_analysing_df = pd.DataFrame(columns = kpi_col)
    
    crmaapplst = []

    for x in monLst:
        if x != 'Annual':
            if  kpi_df[x].sum() <= 20000:
                emp_min = np.random.randint(60, 75)
                crmaapplst.append(emp_min)
            elif kpi_df[x].sum() > 20000 and kpi_df[x].sum() <= 30000:
                emp_min = np.random.randint(70, 100)
                crmaapplst.append(emp_min)
            else:
                emp_min = np.random.randint(90, 120)
                crmaapplst.append(emp_min)
        else:

            if  kpi_months_tot.sum().sum() <= 350000:
                emp_min = np.random.randint(60, 75)
                crmaapplst.append(emp_min)
            elif kpi_months_tot.sum().sum() > 350000 and kpi_months_tot.sum().sum() <= 400000:
                emp_min = np.random.randint(70, 80)
                crmaapplst.append(emp_min)
            else:
                emp_min = np.random.randint(80, 90)
                crmaapplst.append(emp_min)

        mtotal = sum(crmaapplst)

    crmmAna_data = {'Jan': crmaapplst[0] , 'Feb': crmaapplst[1], 'Mar': crmaapplst[2], 'Apr': crmaapplst[3], 
               'May': crmaapplst[4], 'Jun': crmaapplst[5], 'Jul': crmaapplst[6], 'Aug': crmaapplst[7],
               'Sep': crmaapplst[8], 'Oct': crmaapplst[9], 'Nov': crmaapplst[10], 'Dec': crmaapplst[11], 
               'Annual': crmaapplst[12], 'Total': mtotal}

    a_analysing_df = a_analysing_df.append(crmmAna_data, ignore_index = True)
    a_analysing_df.to_csv(a_analysis_csv + '.csv', index=False)
    
    
```

```python
def CRM_employee_dta (crm_1_kpi_csv, crm_2_kpi_csv, crm_3_kpi_csv):
    
     
    import pandas as pd
    import numpy as np
    
    emp_csv = pd.read_csv('emp.csv')
    
    '''
    Efficiency Limitation 
    Age 20-30(30-60%)
    Age 31-50(20-30%)
    Age 51-55(20-25%)

    The AI component helps reduce inconsistencies of manual processing.

    '''
    
    
    '''
    ==================================================================
    CALCULATING KPI EFFICIENCY 

    ==================================================================

    '''
    
    defaultdf = pd.read_csv('emp.csv')
    
    cemp_eff1 = []
    cemp_eff2 = []
    cemp_eff3 = []

    for x in defaultdf.age:
        if x >= 20 and x <= 30:
            y = np.random.randint(30,60)
            cemp_eff1.append(y * .01)
        elif x >= 31 and x <= 50:
            y = np.random.randint(20,30)
            cemp_eff1.append(y * .01)
        else:
            y = np.random.randint(20,25)
            cemp_eff1.append(y * .01)

    for x in defaultdf.age:
        if x >= 20 and x <= 30:
            y = np.random.randint(40,50)
            cemp_eff2.append(y * .01)
        elif x >= 31 and x <= 50:
            y = np.random.randint(25,35)
            cemp_eff2.append(y * .01)
        else:
            y = np.random.randint(15,25)
            cemp_eff2.append(y * .01)

    for x in defaultdf.age:
        if x >= 20 and x <= 30:
            y = np.random.randint(35,50)
            cemp_eff3.append(y * .01)
        elif x >= 31 and x <= 50:
            y = np.random.randint(35,40)
            cemp_eff3.append(y * .01)
        else:
            y = np.random.randint(30, 40)
            cemp_eff3.append(y * .01)

    ckpi_data = ['employeeid','Jan', 'Feb', 'Mar', 'Apr', 
     'May', 'Jun', 'Jul', 'Aug',
     'Sep', 'Oct', 'Nov', 'Dec', 'Avg', 'Total', 'Category']

    ckpi_df1 = pd.DataFrame(columns = ckpi_data)
    ckpi_df2 = pd.DataFrame(columns = ckpi_data)
    ckpi_df3 = pd.DataFrame(columns = ckpi_data)

    '''
    ==================================================================
    CALCULATING 1ST CRM KPI 

    ==================================================================

    '''
    
    
    emp_id_len = len(defaultdf)
    
    for x in range (emp_id_len):
        capplst = []
        for y in range (12):
            if y == 0:
                capp = np.random.randint(168, 200)
                cappeff = (capp * cemp_eff1[x])
                ctot1 = np.ceil(capp + cappeff).astype(int)
                capplst.append(ctot1)
            else:
                capp = np.random.randint(336, 672)
                cappeff = (capp * cemp_eff1[x])
                ctot2 = np.ceil(capp + cappeff).astype(int)
                capplst.append(ctot2)

        caverage = np.ceil (sum(capplst) / 12).astype(int)
        ctotal = sum(capplst)

        ckpi_data1 = {'employeeid': defaultdf.employeeid[x],'Jan': capplst[0] , 'Feb': capplst[1], 'Mar': capplst[2], 'Apr': capplst[3], 
               'May': capplst[4], 'Jun': capplst[5], 'Jul': capplst[6], 'Aug': capplst[7],
               'Sep': capplst[8], 'Oct': capplst[9], 'Nov': capplst[10], 'Dec': capplst[11], 
               'Avg': caverage, 'Total': ctotal, 'Category': 'CRM SYSTEM 1'}

        ckpi_df1 = ckpi_df1.append(ckpi_data1, ignore_index = True)
    ckpi_df1.to_csv(crm_1_kpi_csv + '.csv', index=False)
    
    '''
    ==================================================================
    CALCULATING 2ND CRM KPI 

    ==================================================================

    '''
    
    for x in range (emp_id_len):
        capplst = []
        for y in range (12):
            if y == 0:
                capp = np.random.randint(168, 200)
                cappeff = (capp * cemp_eff2[x])
                ctot1 = np.ceil(capp + cappeff).astype(int)
                capplst.append(ctot1)
            else:
                capp = np.random.randint(336, 672)
                cappeff = (capp * cemp_eff2[x])
                ctot2 = np.ceil(capp + cappeff).astype(int)
                capplst.append(ctot2)

        caverage = np.ceil (sum(capplst) / 12).astype(int)
        ctotal = sum(capplst)

        ckpi_data2 = {'employeeid': defaultdf.employeeid[x],'Jan': capplst[0] , 'Feb': capplst[1], 'Mar': capplst[2], 'Apr': capplst[3], 
               'May': capplst[4], 'Jun': capplst[5], 'Jul': capplst[6], 'Aug': capplst[7],
               'Sep': capplst[8], 'Oct': capplst[9], 'Nov': capplst[10], 'Dec': capplst[11], 
               'Avg': caverage, 'Total': ctotal, 'Category': 'CRM SYSTEM 2'}

        ckpi_df2 = ckpi_df2.append(ckpi_data2, ignore_index = True)
    ckpi_df2.to_csv(crm_2_kpi_csv + '.csv', index=False)
    
    '''
    ==================================================================
    CALCULATING 3RD CRM KPI 

    ==================================================================

    '''
    
    for x in range(emp_id_len):
        capplst = []
        for y in range (12):
            if y == 0:
                capp = np.random.randint(168, 200)
                cappeff = (capp * cemp_eff3[x])
                ctot1 = np.ceil(capp + cappeff).astype(int)
                capplst.append(ctot1)
            else:
                capp = np.random.randint(336, 672)
                cappeff = (capp * cemp_eff3[x])
                ctot2 = np.ceil(capp + cappeff).astype(int)
                capplst.append(ctot2)

        caverage = np.ceil (sum(capplst) / 12).astype(int)
        ctotal = sum(capplst)

        ckpi_data3 = {'employeeid': defaultdf.employeeid[x],'Jan': capplst[0] , 'Feb': capplst[1], 'Mar': capplst[2], 'Apr': capplst[3], 
               'May': capplst[4], 'Jun': capplst[5], 'Jul': capplst[6], 'Aug': capplst[7],
               'Sep': capplst[8], 'Oct': capplst[9], 'Nov': capplst[10], 'Dec': capplst[11], 
               'Avg': caverage, 'Total': ctotal, 'Category': 'CRM SYSTEM 3'}

        ckpi_df3 = ckpi_df3.append(ckpi_data3, ignore_index = True)
    ckpi_df3.to_csv(crm_3_kpi_csv + '.csv', index=False)
    
```

```python
def calc_null_hypothesis (two_mean_size):

    from scipy.stats import ttest_ind as t_test
    import random as ran
    import pandas as pd
    import numpy as np
    
    stats = pd.read_csv('stat.csv')


    upper = stats[stats['rank'] == 1]
    lower = stats[stats['rank'] == 2]

    upper = upper.reset_index(drop = True).drop(columns = ['csv_name','set', 'rank', 'std'])
    lower = lower.reset_index(drop = True).drop(columns = ['csv_name','set', 'rank', 'std'])

    upperlist = []
    lowerlist = []

    for x in upper['mean'].values:
        upperlist.append(x)

    for x in lower['mean'].values:
        lowerlist.append(x)

    ran.shuffle(upperlist)
    ran.shuffle(lowerlist)

    uppersample = []
    lowersample = []

    for x in range (two_mean_size):
        uppersample.append(upperlist[x])
        lowersample.append(lowerlist[x])

    upperarray = np.array(uppersample)
    lowerarray = np.array(lowersample)
    alpha = 0.05

    stat, p = t_test(upperarray, lowerarray, equal_var=False)
    
    
    if p < alpha:
        print ('Based on ' + str(two_mean_size) + ' samples, the P-Value is ' + str(p))
        print ('Sample Set 1: ' + str(uppersample))
        print ('Sample Set 2: ' + str(lowersample))
        print (str(p) + ' is less than 0.05 (Alpha)')
        print ('The null hypothesis must be rejected.')
    else:
        print('Based on ' + str(two_mean_size) + ' samples, the P-Value is ' + str(p))
        print ('Sample Set 1: ' + str(uppersample))
        print ('Sample Set 2: ' + str(lowersample))
        print (str(p) + ' is equal or greater than 0.05 (Alpha)')
        print ('The null hypothesis must be accepted.')
```
